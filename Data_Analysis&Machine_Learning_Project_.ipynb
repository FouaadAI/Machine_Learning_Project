### *1. Data Cleaning*

Before diving into analysis, it's crucial to clean the data.

```python
import pandas as pd

# Load the data
df = pd.read_csv('Crime_Incidents_in_2024.csv')
df
```

```python
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
#Drop Columns
df.drop(columns=["OCTO_RECORD_ID", "OBJECTID"], inplace=True)
```

```python
# Convert date columns to datetime
df['REPORT_DAT'] = pd.to_datetime(df['REPORT_DAT'])
df['START_DATE'] = pd.to_datetime(df['START_DATE'])
df['END_DATE'] = pd.to_datetime(df['END_DATE'])
```

```python
# Drop rows with missing critical data (e.g., LATITUDE, LONGITUDE)
df = df.dropna(subset=['LATITUDE', 'LONGITUDE', 'OFFENSE', 'SHIFT', 'METHOD','END_DATE','END_DATE'])
```

```python
# Fill missing categorical values with "Unknown"
categorical_cols = ["WARD",
                    "ANC",
                    "DISTRICT",
                    "PSA",
                    "NEIGHBORHOOD_CLUSTER",
                    "BLOCK_GROUP",
                    "CENSUS_TRACT",
                    "VOTING_PRECINCT",
                    "BID"]
for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
df.info()
```

### *2. Exploratory Data Analysis (EDA)*

#### *Question 1: What are the most common types of crimes?*

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each crime type
crime_counts = df['OFFENSE'].value_counts()
crime_counts
```

```python
# Plot the top 10 most common crimes
plt.figure(figsize=(10, 6))
crime_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Common Crimes')
plt.xlabel('Crime Type')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.show()
```

*Visualization:*
Top 10 Most Common Crimes
The bar chart shows that "THEFT/OTHER" and "THEFT F/AUTO" are the most common crimes.

#### *Question 2: How are crimes distributed by shift and method?*

We can analyze the distribution of crimes by SHIFT (DAY, EVENING, MIDNIGHT) and METHOD (GUN, KNIFE, OTHERS).

```python
# Crime distribution by shift
shift_counts = df['SHIFT'].value_counts()
shift_counts
```

```python
# Crime distribution by method
method_counts = df['METHOD'].value_counts()
method_counts
```

```python
# Plotting
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

shift_counts.plot(kind='bar', ax=axes[0], color='lightgreen')
axes[0].set_title('Crime Distribution by Shift')
axes[0].set_xlabel('Shift')
axes[0].set_ylabel('Number of Incidents')

method_counts.plot(kind='bar', ax=axes[1], color='salmon')
axes[1].set_title('Crime Distribution by Method')
axes[1].set_xlabel('Method')
axes[1].set_ylabel('Number of Incidents')

plt.tight_layout()
plt.show()
```

*Visualization:*
Crime Distribution by Shift and Method
The bar charts show that most crimes occur during the "EVENING" AND "DAY" shift, and the most common method is "OTHERS."

#### *Question 3: Where are the crime hotspots?*

We can use geospatial analysis to identify crime hotspots by plotting the crimes on a map.

```python
import plotly.express as px
```

```python
# ---- 1. SCATTERPLOT OF CRIME LOCATIONS ----
plt.figure(figsize=(10, 6))
plt.scatter(df["LONGITUDE"], df["LATITUDE"], alpha=0.3, color="red", s=5)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Crime Hotspots (Scatterplot)")
plt.grid(True)
plt.show()

# ---- 2. HEATMAP OF CRIME DENSITY ----
plt.figure(figsize=(10, 6))
sns.kdeplot(x=df["LONGITUDE"], y=df["LATITUDE"], cmap="Reds", fill=True, alpha=0.5)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Crime Density Heatmap")
plt.show()

# ---- 3. INTERACTIVE HEATMAP USING PLOTLY ----
fig = px.density_mapbox(df, lat="LATITUDE", lon="LONGITUDE", radius=10, opacity=0.5,
                        center={"lat": df["LATITUDE"].median(), "lon": df["LONGITUDE"].median()},
                        zoom=10, mapbox_style="open-street-map",
                        title="Interactive Crime Hotspot Heatmap")
fig.show()
```

*Visualization:*
Crime Hotspots Map
The heatmap shows that the downtown area and certain neighborhoods are crime hotspots.

#### *Question 4: Are there any trends in crime over time?*

We can analyze the number of crimes reported over time to identify trends.

```python
# Group by date and count the number of crimes
df['REPORT_DATE'] = df['REPORT_DAT'].dt.date
crime_trend = df.groupby('REPORT_DATE').size()
crime_trend
```

```python
# Plot the trend
plt.figure(figsize=(12, 6))
crime_trend.plot(kind='line', color='purple')
plt.title('Crime Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.grid(True)
plt.show()
```

*Visualization:*
Crime Trends Over Time  
The line chart shows a slight increase in crimes during certain months, possibly indicating seasonal trends.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity
from sklearn.model_selection import GridSearchCV

# Load coordinates data (assuming it's in a CSV file)
df= pd.read_csv("crim_analyzed.csv")  # Replace with your file path
coordinates = df[["LONGITUDE", "LATITUDE"]].dropna()
# Optimize bandwidth for KDE
params = {"bandwidth": np.linspace(0.001, 0.01, 20)}
grid = GridSearchCV(KernelDensity(kernel="gaussian"), params)
grid.fit(coordinates)

best_kde = grid.best_estimator_
print(f"Best bandwidth: {grid.best_params_['bandwidth']}")

# Create a grid covering the geographic area
xmin, xmax = coordinates["LONGITUDE"].min(), coordinates["LONGITUDE"].max()
ymin, ymax = coordinates["LATITUDE"].min(), coordinates["LATITUDE"].max()

xx, yy = np.meshgrid(np.linspace(xmin, xmax, 100), np.linspace(ymin, ymax, 100))
grid_points = np.vstack([xx.ravel(), yy.ravel()]).T

# Calculate density scores for the grid
log_density = best_kde.score_samples(grid_points)
density = np.exp(log_density).reshape(xx.shape)

# Threshold for high-risk areas (top 10% density)
threshold = np.percentile(density, 90)
high_risk = np.where(density > threshold)

# Get coordinates of high-risk areas
high_risk_coords = grid_points[high_risk[0]]
print(f"High-risk areas (longitude, latitude):\n{high_risk_coords}")

# Ask user for input method
input_method = input("Do you want to input a new observation manually or from a file? (manual/file): ").strip().lower()

if input_method == "manual":
    # Prompt user for new observation
    new_longitude = float(input("Enter the longitude of the new observation: "))
    new_latitude = float(input("Enter the latitude of the new observation: "))
    new_observation = np.array([[new_longitude, new_latitude]])  # Longitude, Latitude
elif input_method == "file":
    # Load new observations from a file
    new_observations_file = input("Enter the path to the file containing new observations: ")
    new_observations = pd.read_csv(new_observations_file)
    new_observation = new_observations[["LONGITUDE", "LATITUDE"]].values
else:
    raise ValueError("Invalid input method. Choose 'manual' or 'file'.")

# Predict the log-density for the new observation(s)
log_density_new = best_kde.score_samples(new_observation)

# Convert log-density to density
density_new = np.exp(log_density_new)

# Check if the new observation(s) is in a high-risk area
for i, (lon, lat) in enumerate(new_observation):
    print(f"Crime Density for observation {i+1}, ({lon}, {lat}): {density_new[i]}")
    if density_new[i] > threshold:
        print("This location is in a high-risk area for crime.")
    else:
        print("This location is not in a high-risk area for crime.")

# Plot the results
plt.figure(figsize=(12, 8))
plt.scatter(coordinates["LONGITUDE"], coordinates["LATITUDE"], s=5, c="red", alpha=0.1)
plt.contourf(xx, yy, density, levels=50, cmap="Reds", alpha=0.6)
plt.colorbar(label="Crime Density")
plt.scatter(new_observation[:, 0], new_observation[:, 1], s=100, c="blue", marker="x", label="New Observation")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Crime Hotspots with New Observation")
plt.legend()
plt.show()
```
